Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
(Hint: Compare their scope, techniques, and applications for each.)
Answer:AI is the broad concept of creating intelligent machines, ML is a subset of AI where systems learn from data, and DL is a subset of ML using deep neural networks to handle complex data. Data Science is a multidisciplinary field that uses a variety of tools, including ML and DL, to extract insights from data to inform decision-making, and it typically involves a human in the loop for analysis and interpretation.
1.Artificial Intelligence (AI)
Scope: The broadest field, focused on creating systems that can perform tasks that typically require human intelligence, such as problem-solving and decision-making.
Techniques: Employs a wide range of techniques, including logic, decision trees, and, more advanced techniques like machine learning and deep learning.
Applications: Chatbots, voice assistants (like Siri or Alexa), and robotics.
2.Machine Learning (ML)
Scope: A subset of AI that focuses on developing algorithms that allow computers to learn from data without being explicitly programmed.
Techniques: Uses statistical models and algorithms to find patterns in data to make predictions or decisions. Examples include regression, classification, and clustering.
Applications: Recommendation systems (like those on Netflix or Spotify), search engine algorithms, and spam filtering.
3.Deep Learning (DL)
Scope: A subset of ML that uses multi-layered artificial neural networks to process vast amounts of data.
Techniques: Mimics the human brain's structure to identify complex patterns, often requiring large datasets for training. It excels with unstructured data like images and text.
Applications: Image recognition, natural language processing (NLP), self-driving cars, and medical diagnosis (like tumor detection).
4.Data Science
Scope: An interdisciplinary field that encompasses the entire process of extracting knowledge and insights from data. It includes data collection, cleaning, analysis, and interpretation.
Techniques: Combines statistics, mathematics, computer science, and domain expertise to analyze data, build predictive models, and communicate findings. It frequently uses ML and DL as tools, but is not limited to them.
Applications: Business intelligence, fraud detection, market forecasting, and helping businesses make informed decisions

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.
Answer:Overfitting and underfitting represent two common challenges in machine learning model development, both leading to poor generalization performance on unseen data.
1. Overfitting:
Explanation: Overfitting occurs when a model learns the training data too well, including the noise and specific patterns, leading to excellent performance on the training set but poor performance on new, unseen data. This is characterized by high variance.
Detection:
Performance Gap: A significant difference between high training accuracy/low training error and significantly lower validation/test accuracy or higher validation/test error.
Learning Curves: Training error decreases significantly, while validation error starts to increase after a certain point.
Prevention:
Regularization: Techniques like L1 (Lasso) and L2 (Ridge) regularization add a penalty to the model's loss function for large coefficient values, encouraging simpler models.
Cross-validation: Using methods like K-fold cross-validation helps assess the model's generalization ability by training and evaluating it on multiple subsets of the data.
Early Stopping: Halting the training process when the model's performance on a validation set starts to degrade, preventing it from memorizing the training data.
Feature Selection/Engineering: Reducing the number of features or creating more meaningful features to reduce noise and redundancy.
Ensemble Methods: Combining multiple models (e.g., Random Forests, Gradient Boosting) to reduce variance and improve generalization.
2. Underfitting:
Explanation: Underfitting occurs when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both the training and test sets. This is characterized by high bias.
Detection:
Low Performance: Low accuracy or high error on both the training and validation/test sets.
Learning Curves: Training error remains high and does not significantly decrease, and the validation error mirrors the training error, also remaining high.
Prevention:
Increase Model Complexity: Using a more complex model architecture (e.g., more layers in a neural network, a more complex algorithm).
Feature Engineering: Creating new features or transforming existing ones to provide the model with more relevant information.
Reducing Regularization: If regularization is being used, reducing its strength to allow the model more flexibility to learn.
Increasing Training Time/Epochs: Allowing the model more time to learn the patterns in the data (though this must be balanced with the risk of overfitting).
3. Bias-Variance Tradeoff:
This fundamental concept highlights the inherent tension between bias and variance in model complexity.
High Bias (Underfitting): A model with high bias makes strong assumptions about the data, leading to a simplified representation and inability to capture the true relationship.
High Variance (Overfitting): A model with high variance is overly sensitive to the training data, capturing noise and specific patterns, leading to poor generalization.

Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.
Answer:You can handle missing values by deleting rows or columns with missing data, imputing values using the mean or median of the column, or using predictive modeling to estimate the missing values based on other features. Deletion is simple but can lose data, while imputation preserves data but can skew distributions, and predictive modeling is more complex but provides more accurate estimates. 1. Deletion Method: Remove rows or columns that contain missing values. This is best when a small percentage of the data is missing and the dataset is large.Example: If you have a dataset of 1,000 customers and only 10 have a missing age, you can delete those 10 rows. If a column like "number of children" is missing for 80% of the customers, you might delete the entire column. 2. Mean/Median Imputation Method: Replace missing values in a column with the mean (average) or median (middle value) of the non-missing values in that same column.Example:Mean: If you have a column for "Income" with values of \(\$50,000\), \(\$60,000\), and \(\$70,000\) (with two missing values), you would calculate the mean as \(\$60,000\). You would then fill the two missing spots with \(\$60,000\).Median: If the "Income" column was skewed (e.g., \(\$50,000\), \(\$55,000\), \(\$200,000\)), the median would be a better representation. If the sorted values are \(\$50,000\), \(\$55,000\), \(\$200,000\), the median is \(\$55,000\), and you would use that to fill the missing values. 3. Predictive Modeling Method: Use other columns in the dataset to train a model that predicts the missing value. This is a more advanced method that can provide more accurate estimations.Example: You can use a regression model (like Linear Regression) to predict the missing "Income" values based on other features like "Education Level" and "Years of Experience." You would train the model on the rows where income is not missing, and then use that trained model to predict the income for the rows where it is missing

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.
Answer:An imbalanced dataset in machine learning refers to a classification problem where the distribution of instances across different classes is not equal. One class, known as the majority class, significantly outnumbers the other class(es), referred to as the minority class(es). This imbalance can lead to models that are biased towards the majority class, resulting in poor performance on the minority class, which is often the class of interest (e.g., fraud detection, disease diagnosis).
Two techniques to handle imbalanced datasets:

Synthetic Minority Over-sampling Technique (SMOTE):
Theoretical: SMOTE is an oversampling technique that generates synthetic samples for the minority class. It works by selecting a minority class instance and then finding its k-nearest minority class neighbors. New synthetic instances are created by interpolating between the selected instance and its neighbors in the feature space. This approach increases the number of minority class samples without simply duplicating existing ones, thereby adding diversity to the dataset.

Class Weights in Models:
Theoretical: This technique involves assigning different weights to the classes during model training. The idea is to penalize misclassifications of the minority class more heavily than misclassifications of the majority class. This forces the model to pay more attention to the minority class instances, even though they are fewer in number, and reduces the bias towards the majority class. Many machine learning algorithms (e.g., Logistic Regression, Support Vector Machines, Gradient Boosting) offer a class_weight parameter.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.
Answer:Feature scaling is a crucial preprocessing step in machine learning to ensure that numerical features in a dataset have a comparable scale. This prevents algorithms from being disproportionately influenced by features with larger magnitudes, leading to more robust and accurate models.                Impact on Algorithms:Distance-based algorithms (e.g., KNN, SVM, K-Means): These algorithms rely on calculating distances between data points. Without scaling, features with larger ranges will dominate the distance calculations, making the algorithm biased towards those features. For example, in KNN, the "nearest" neighbors might be determined more by a feature with a large range (e.g., income) than by a feature with a smaller range (e.g., age), even if age is more relevant. Scaling ensures all features contribute equally to the distance metric. Gradient Descent based algorithms (e.g., Linear Regression, Logistic Regression, Neural Networks): Gradient descent optimizes model parameters by iteratively moving in the direction of the steepest descent of the loss function. If features are on vastly different scales, the loss function's contours will be elongated and narrow, leading to oscillations and slower convergence for gradient descent. Scaling creates a more spherical loss surface, allowing for faster and more stable convergence.      
Min-Max Scaling vs. Standardization:     Min-Max Scaling (Normalization): Transforms features to a specific range, typically between 0 and 1. Formula: \(X_{scaled}=\frac{X-X_{min}}{X_{max}-X_{min}}\) Advantages: Simple to understand and implement, preserves the original distribution shape (except for scaling), useful when features need to be within a specific bounded range. Disadvantages: Highly sensitive to outliers, as \(X_{min}\) and \(X_{max}\) are directly affected by extreme values. Standardization (Z-score Normalization): Transforms features to have a mean of 0 and a standard deviation of 1.  Formula: \(X_{scaled}=\frac{X-\mu }{\sigma }\) (where \(\mu \) is the mean and \(\sigma \) is the standard deviation) Advantages: Less affected by outliers than Min-Max scaling, useful for algorithms that assume a Gaussian distribution, and effective for features with no clear upper or lower bounds. Disadvantages: Does not bound features to a specific range, which might be undesirable for some algorithms or visualizations.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.
Answer:Label Encoding assigns an integer to each category, best for ordinal data where order matters, while One-Hot Encoding creates a new binary column for each category, best for nominal data where there is no order. You would prefer Label Encoding for ordinal data to preserve the inherent order, whereas you would prefer One-Hot Encoding for nominal data to avoid creating a false hierarchical relationship between categories. However, if you have a high number of unique nominal categories, Label Encoding is more memory-efficient, but One-Hot Encoding is generally preferred for nominal data to avoid misinterpreting the numerical values as having a rank.
When to use which
Use Label Encoding for:
Categories with a natural, meaningful order (e.g., "low," "medium," "high" or "small," "medium," "large").
Situations where dimensionality needs to be kept low and the order is not a problem, such as with certain tree-based models.
Use One-Hot Encoding for:
Categories without any inherent order (e.g., city names, colors).
When you want to avoid introducing a false sense of order that might bias the model.

Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)
Answer:

```python
import pandas as pd
import requests

# URL of the raw titanic.csv file on GitHub
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"

# Download the CSV file
response = requests.get(url)
response.raise_for_status()  # Raise an exception for bad status codes
with open("titanic.csv", "wb") as f:
    f.write(response.content)

# Load the Titanic dataset from the downloaded file
df = pd.read_csv("titanic.csv")

# Calculate survival rates by Pclass
survival_by_class = df.groupby("Pclass")["Survived"].mean()
print("Survival rates by Pclass:")
print(survival_by_class)
print("\n")

# Group passengers into children and adults
df["Child"] = df["Age"] < 18
survival_by_age_group = df.groupby("Child")["Survived"].mean()
print("Survival rates by age group (Child < 18):")
print(survival_by_age_group)
```
